```python
# filter for waterfront (wf), look at stats for prices and grades
df_wf = df.query('waterfront==1' )
df_nwf = df.query('waterfront==0')
print('no waterfront houses:')
display(df_nwf[['price', 'grade']].describe())
print('waterfront houses:')
display( df_wf[['price', 'grade']].describe())
```

at first glance, when looking at the means, houses without waterfront on average cost less (532286 +/- 342808) than houses with waterfront (1717214 +/- 1145384).

the grades for houses without waterfront is on average 7.7 +/- 1.2 and waterfront houses are higher graded with 8.8 +/- 1.8 on average.

Testing if these differences are significant:

```python
#pip install scipy 
from scipy import stats

#comparing grades of waterfront(wf) and non waterfront (nwf) houses
#one sided t-test. 
# H0 : µ_wf <= µ_nwf         H1: µ_wf > µ_nwf

alpha = 0.05  
t_stat_g, p_value_g = stats.ttest_ind(df_wf['grade'] , df_nwf['grade'], alternative='greater')
# results
print('t-statistic:', t_stat_g)
print('p-value:', p_value_g)
if p_value_g < alpha:
    print('reject H0: waterfront houses have significantly higher grades')
else:
    print('not reject H0: waterfront houses have not significantly higher grades')
```

```python
#same for prices
#one sided t-test. 
# H0 : µ_nf <= µ_nwf         H1: µ_wf > µ_nwf

alpha = 0.05  
t_stat_p, p_value_p = stats.ttest_ind(df_wf['price'] , df_nwf['price'], alternative='greater')
# results
print('t-statistic:', t_stat_p)
print('p-value:', p_value_p)
if p_value_p < alpha:
    print('reject H0: waterfront houses are significantly more expensive')
else:
    print('not reject H0: waterfront houses are not significantly more expensive')
```

results: waterfront houses have significantly higher grades and prices!

### Hypothesis 2: houses with high sqft_living have also nighbours with high sqft_living 

- sqft_living ~ sqft_living15 ?
- correlation?
- makes it hard to find houses with higher sqft_living  than the surrounding houses


```python
# check correlation
mat_h2= df[ [ "sqft_living", "sqft_living15"]].corr()
display(mat_h2)
```

--> correlation between sqft_living and the sqd´ft_living of the 15 surrounding neighboours is quite high with 0,756.

on the other hand, one could have expected a higher correlation , maybe even close to perfect (1)


is this correlation significant?

```python
from scipy.stats import pearsonr

#pearson correlation coefficient and the p-value
# H0 : no correlation       H1: correlation
corr_sq, p_value_pear = pearsonr(df['sqft_living'], df['sqft_living15'])

print('pearson correlation coefficient:', corr_sq)
print('p-value:', p_value_pear)

#the results
alpha = 0.05
if p_value_pear < alpha:
    print('H0 rejected: correlation is statistically significant')
else:
    print(' correlation is not statistically significant')
```

the correlation is statistically significant!


Create a Scatterplot with linear regression to visually show correlation.

link between slope of linear regression is correlation coefficient times the quotient of standard deviations:
 
 b = corr_sq * std(sqft_living15) / std(sqft_living)

```python
#scatterplot with linear regression 

fig3, ax3 = plt.subplots(figsize=(8,8))

h3 = ax3.scatter(x=df['sqft_living'], y=df['sqft_living15'])

# lin. regression = 1. grade polynom
b3, a3 = np.polyfit(x=df['sqft_living'], y=df['sqft_living15'], deg=1)

# plot regression lines
ax3.plot(df['sqft_living'], a3 + b3 * df['sqft_living'], color="k", lw=2)

#adding labels and some formating
ax3.set_title('sqft_living vs. sqft_living15')
ax3.set_xlabel('sqft_living')
ax3.set_ylabel('sqft_living15')
ax3.set_xlim(right=8000, left=0)
ax3.set_ylim(top=8000, bottom=0)

#adding formula for lin. fit line
ax3.text(1000, 7500, 'y = a + b * x', fontsize=12)
ax3.text(1000, 7200, 'a = '+str(round(a3,2)), fontsize=12)
ax3.text(1000, 6900, 'b= '+str(round(b3,2)), fontsize=12)
```

- scatterplot also shows positive correlation
- conclusion: houses with higher sqft_living indeed have neighbours with higher sqft_living !

### Hypothesis 3:
### houses with twice as much sqft_living than their neighbours are much more expensive

- Group "showoff houses" : sqft_living > 2 * sqft_living15
- Group "normal houses" : sqft_living <= 2 * sqft_living15

```python
# cerating column for price per sqft_living
df['pr_sqft_l'] = df['price'] / df['sqft_living']

#creating the 2 groups
df_norm = df.query('sqft_living <= 2 * sqft_living15')
print('normal houses:')
display(df_norm[['price', 'pr_sqft_l', 'sqft_living', 'sqft_living15']].describe())

print('showoff houses:')
df_showoff = df.query('sqft_living > 2 * sqft_living15')
display(df_showoff[['price', 'pr_sqft_l', 'sqft_living', 'sqft_living15']].describe())
```

first i want to compare the overall pricees 

```python
# one sided t-test to compare the means of the two groups
# H0 : µ_s <= µ_n         H1: µ_s > µ_n

t_stat_show, p_value_show = stats.ttest_ind(df_showoff['price'], df_norm['price'], equal_var=False, alternative='greater')
alpha = 0.05

print('t-statistic:', t_stat_show)
print('p-value:', p_value_show)
if p_value_show < alpha:
    print('reject H0: showoff houses are significantly more expensive')
else:
    print('not reject H0: showoff houses are not significantly more expensive')
```

conclusion: houses with twice as much sqft_living than their 15 neighbours are significantly more expensive

```python
# boxplots for both groups
fig_boxx, axxes = plt.subplots(1, 2, figsize=(8, 6))

boxx_norm = sns.boxplot(data=df_norm, y='price', ax=axxes[0])
boxx_showoff = sns.boxplot(data=df_showoff, y='price', ax=axxes[1], color='orange')

axxes[0].set_title('normal houses')
axxes[0].set_ylabel('price')
axxes[1].set_title('showoff houses')
axxes[1].set_ylabel('price')
axxes[0].set_ylim(top=8000000)
axxes[1].set_ylim(top=8000000)
plt.show()
```



is this also the case for the price per sqft_living?

```python
# one sided t-test to compare the means of the two groups
# H0 : µ_s <= µ_n         H1: µ_s > µ_n

t_stat_psqftl, p_value_psqftl = stats.ttest_ind(df_showoff['pr_sqft_l'], df_norm['pr_sqft_l'], alternative='greater')
alpha = 0.05

print('t-statistic:', t_stat_psqftl)
print('p-value:', p_value_psqftl)
if p_value_psqftl < alpha:
    print('reject H0: showoff houses have significantly higher prices per sqft_living')
else:
    print('not reject H0: showoff houses do not have significantly higher prices per sqft_living')
```

no, the price per sqft_living is not significantly higher for showoff houses.

is there even a difference?

```python
# two sided test:
# H0: µ_n = µ_s
# H1 : µ_n != µ_s

t_stat_psqftl2, p_value_psqftl2 = stats.ttest_ind(df_showoff['pr_sqft_l'], df_norm['pr_sqft_l'],)
alpha = 0.05

print('t-statistic:', t_stat_psqftl2)
print('p-value:', p_value_psqftl2)
if p_value_psqftl2 < alpha:
    print('reject H0: there is a significant difference in price per sqft_living between showoff houses and normal houses')
else:
    print('not reject H0: there is NO significant difference in price per sqft_living between showoff houses and normal houses')
```

ok, so maybe even a lower price?

```python
# one sided t-test to compare the means of the two groups
# H0 : µ_s >= µ_n         H1: µ_s < µ_n

t_stat_psqftl3, p_value_psqftl3 = stats.ttest_ind(df_showoff['pr_sqft_l'], df_norm['pr_sqft_l'], alternative='less')
alpha = 0.05

print('t-statistic:', t_stat_psqftl3)
print('p-value:', p_value_psqftl3)
if p_value_psqftl3 < alpha:
    print('reject H0: showoff houses have significantly lower prices per sqft_living')
else:
    print('not reject H0: showoff houses do not have significantly lower prices per sqft_living')
```
